```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
spearman_df = pd.read_csv("aae_spearman_results.csv")
spearman_df.head()
```

```python
kendall_df = pd.read_csv("aae_kendall_results.csv")
kendall_df.head()
```

```python
rankings_diff_df = pd.read_csv("aae_rankings_diff_results.csv")
rankings_diff_df.head()
```

```python
top_ten_df = pd.read_csv("aae_top_ten_results.csv")
top_ten_df.head()
```

```python
datasets = list(spearman_df["Dataset"].to_numpy())
percents = [10, 20, 30, 40, 50, 60, 70, 80, 90]

spearman_arr = spearman_df.to_numpy()

for i, row in enumerate(spearman_arr):
    plt.plot(percents, row[1:], label=datasets[i])

plt.ylim(0, 1)
plt.legend()
```

```python
kendall_arr = kendall_df.to_numpy()

for i, row in enumerate(kendall_arr):
    plt.plot(percents, row[1:], label=datasets[i])

plt.ylim(0, 1)
plt.legend()
```

```python
top_ten_arr = top_ten_df.to_numpy()

for i, row in enumerate(top_ten_arr):
    plt.plot(percents, row[1:], label=datasets[i])

plt.legend()
```

```python
rankings_diff_arr = rankings_diff_df.to_numpy()

for i, row in enumerate(rankings_diff_arr):
    plt.plot(percents, row[1:], label=datasets[i])

plt.legend()
plt.ylim(0, 100)
```
